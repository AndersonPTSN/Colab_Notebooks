# Calculadora digital
Para demonstrar os conhecimentos obtidos, foi criada uma interface para interagir com as principais funções de uma calculadora padrão, como:
* soma
* subtração
* divisão
* multiplicação

Além de outras funções, tais como:

* Recíproco (1/x)
* Quadrado (x^2)
* Raiz Quadrada (raiz de x)

Juntamente com a biblioteca math.

```python
import math

#Verificação----------------------------------
def isnumber(value):
    try:
         float(value)
    except ValueError:
         return False
    return True

#Funções----------------------------------
def soma():
  valor1 = input("Digite o primeiro valor a ser somado: ")
  if(isnumber(valor1)):
    valor2 = input("Digite o segundo valor a ser somado: ")
    if(isnumber(valor2)):
      print("Resultado: ",float(valor1) + float(valor2))
    else:
      print("Valor Inválido")
  else:
    print("Valor Inválido")

def subtracao():
  valor1 = input("Digite o primeiro valor a ser subtraido: ")
  if(isnumber(valor1)):
    valor2 = input("Digite o segundo valor a subtrair: ")
    if(isnumber(valor2)):
      print("Resultado: ",float(valor1) - float(valor2))
    else:
      print("Valor Inválido")
  else:
    print("Valor Inválido")

def divisao():
  valor1 = input("Digite o valor do dividendo: ")
  if(isnumber(valor1)):
    valor2 = input("Digite o valor do divisor: ")
    if(isnumber(valor2)):
      if(float(valor2) != 0):
        print("Resultado: ",float(valor1) / float(valor2))
      else:
        print("Divisão por zero, sem resultado")
    else:
      print("Valor Inválido")
  else:
    print("Valor Inválido")

def multiplicacao():
  valor1 = input("Digite o primeiro valor a ser multiplicado: ")
  if(isnumber(valor1)):
    valor2 = input("Digite o segundo valor a multiplicar: ")
    if(isnumber(valor2)):
      print("Resultado: ",float(valor1) * float(valor2))
    else:
      print("Valor Inválido")
  else:
    print("Valor Inválido")

def div():
  valor1 = input("Digite o valor do divisor: ")
  if(isnumber(valor1)):
    if(float(valor1) != 0):
      print("Resultado: ",1 / float(valor1))
    else:
      print("Divisão por zero, sem resultado")
  else:
    print("Valor Inválido")

def sqr():
  valor1 = input("Digite o valor: ")
  if(isnumber(valor1)):
    print("Resultado: ",float(valor1) * float(valor1))
  else:
    print("Valor Inválido")

def raiz():
  valor1 = input("Digite o valor: ")
  if(isnumber(valor1)):
    if(float(valor1) >= 0):
      print("Resultado: ",math.sqrt(float(valor1)))
    else:
      print("Função não permite numeros negativos")
  else:
    print("Valor Inválido")

#Interface---------------------------------------------
def man():
  while(1 == 1):
    print("Escolha sua opção:")
    print("1 - soma")
    print("2 - subtração")
    print("3 - divisão")
    print("4 - multiplicação")

    print("5 - 1/x")
    print("6 - x^2")
    print("7 - raiz de x")

    print("0 - sair")
    menu = input()

    if menu == "1":
      soma()
    elif menu == "2":
      subtracao()
    elif menu == "3":
      divisao()
    elif menu == "4":
      multiplicacao()

    elif menu == "5":
      div()
    elif menu == "6":
      sqr()
    elif menu == "7":
      raiz()


    elif menu == "0":
      break

man()

```
