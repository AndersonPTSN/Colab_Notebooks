# Agenda telefônica digital
Para mostrar os conhecimentos obtidos foi criado um CRUD de uma agenda telefônica na qual tem as opções de Criar, Ler, Atualizar e Apagar um contato, utilizando conhecimentos de classe juntamente com a biblioteca Sys




```
#Imports----------------------------------
from sys import exit

#Variaveis globais----------------------------------
Agenda = []

#Classes----------------------------------
class Contato():
  nome = ""
  telefone = ""
  def __init__(self,nome,telefone):
    self.nome = nome
    self.telefone = telefone

#Funções----------------------------------
def Criar():
  print("Digite o nome do contato:")
  nome = input()
  print("Digite o número do contato:")
  telefone = input()
  contato = Contato(nome,telefone)
  Agenda.append(contato)
  print("Contato criado com sucesso!")

def Listar():
  if(Agenda == []):
    print("Nenhum contato cadastrado")
  else:
    print("CONTATOS:")
    for contato in Agenda:
      print("NOME: "+contato.nome)
      print("TELEFONE: "+contato.telefone)
      print("-------------------")

def Deletar():
  if(Agenda == []):
    print("Nenhum contato cadastrado")
  else:
    Listar()
    nome = input("Digite o nome do contato no qual voce deseja apagar:")
    for contato in Agenda:
      if(contato.nome == nome):
        Agenda.remove(contato)
        print("Contato "+nome+" removido com sucesso!")
        nome = "@-231DC1GHFDfdg___f:f..;"
        break
    if(nome != "@-231DC1GHFDfdg___f:f..;"):
      print("Contato "+nome+" não encontrado")

def Editar():
  if(Agenda == []):
    print("Nenhum contato cadastrado")
  else:
    Listar()
    nome = input("Digite o nome do contato no qual voce deseja editar:")
    for contato in Agenda:
      if(contato.nome == nome):
        indice = Agenda.index(contato)
        nome = input("Digite o nome do contato: (caso não queira editar, deixe o campo em branco)")
        telefone = input("Digite o número do contato: (caso não queira editar, deixe o campo em branco)")
        if(nome != ""):
          Agenda[indice].nome = nome
        if(telefone != ""):
          Agenda[indice].telefone = telefone
        print("Contato "+nome+" editato com sucesso!")
        nome = "@-231DC1GHFDfdg___f:f..;"
        break
    if(nome != "@-231DC1GHFDfdg___f:f..;"):
      print("Contato "+nome+" não encontrado")

#Interface---------------------------------------------
while(1):
  print("")
  print("")
  print("AGENDA TELEFONICA DIGITAL, digite o número da opção que voce deseja:")
  print("1 - Criar contatos")
  print("2 - Listar contato")
  print("3 - Editar contato")
  print("4 - Deletar contato")
  print("0 - Sair")

  menu = input()

  if(menu in {"1","2","3","4","0"}):
    if(menu == "1"):
      Criar()
    if(menu == "2"):
      Listar()
    if(menu == "3"):
      Editar()
    if(menu == "4"):
      Deletar()
    if(menu == "0"):
      exit()

  else:
    print("Por favor, escolha uma opção válida")

```
